```python
"""

  4. Write a program using an object oriented approach that reads a CSV file containing
      information about different regular convex polygons (i.e., all sides equal, all angles equal, lines
      do not cross), and circles. Each row of the file contains two columns, the first is the name of the
      shape, the second is the shape side length or radius. The program should output text describing
      the perimeter and area of each shape in the CSV file, or an error if the input data is invalid.
      Your program should accept the following shapes:
      • triangle
      • square
      • pentagon
      • circle
      The following represents an example CSV file the program should handle:
      triangle,3.5
      circle,2
      Given the input above, your program should output:
      A triangle with side length 3.5 u has a perimeter of 10.5 u and an area of 5.30 u^2
      A circle with radius 2 u has a perimeter of 12.57 u and an area of 12.57 u^2
      Along with the program code, include a separate CSV file where each row represents a different
      test case the program can handle. Be thorough when itemizing these test cases for exercising
      the above requirements.


"""
import math
class Triangle(Polygon):
  def __init__(self,value):
    super().__init__("triangle","side length",value)

  def perimeterFunc(self):
    self.perimeter = 3 * self.side
    return self.perimeter

  def areaFunc(self):
    self.area = (math.sqrt(3) ** self.side)/4
    return  self.area

 #--------
class Circle(Polygon):
    def __init__(self,value):
      super().__init__("circle","radius",value)
      self.pi = 3.14
    
    def perimeterFunc(self):
      self.perimeter = 2 * self.pi * self.side
      return self.perimeter

    def areaFunc(self):
      self.area = self.pi * self.side * self.side
      return  self.area

#---------
class Square(Polygon):
  def __init__(self,value):
    super().__init__("square","side length",value)

  def perimeterFunc(self):
    self.perimeter = 4 * self.side
    return self.perimeter

  def areaFunc(self):
    self.area = self.side * self.side
    return  self.area
#-----------
class Pentagon(Polygon):
  def __init__(self,value):
    super().__init__("pentagon","side length",value)

  def perimeterFunc(self):
    self.perimeter = 5 * self.side
    return self.perimeter

  def areaFunc(self):
    self.area = 0.25 * math.sqrt(5 * (5 + (2*math.sqrt(5)))) * self.side * self.side
    return  self.area
#----------------


import pandas as pd
class Polygon:
  def __init__(self,shape,side,length):
    self.currentShape = shape
    self.currentSide = side
    self.perimeter = None
    self.area = None
    self.side = length

  def showResult(self):
    print("A ", self.currentShape," with ", self.currentSide, self.side , " u has a perimeter of ",self.perimeter," u and an area of ",self.area," u^2")

if __name__ == "__main__":
  #we don’t want to read all lines into a list of lists and then iterate over it, because that will not be an efficient solution for large csv file i.e. file with size in GBs. We are looking for solutions where we read & process only one line at a time while iterating through all rows of csv, so that minimum memory is utilized.
  inputURL = "https://raw.githubusercontent.com/maahidg03/ERezLifeAssignment/main/eRezLifeInput.csv"
  input = pd.read_csv(inputURL,header=None)
  for index, row in input.iterrows():
    if row[0] == "triangle":
      tri = Triangle(row[1])
      tri.perimeterFunc()
      tri.areaFunc()
      tri.showResult()
    elif row[0] == "circle":
      circ = Circle(row[1])
      circ.perimeterFunc()
      circ.areaFunc()
      circ.showResult()
    elif row[0] == "square":
      squ = Square(row[1])
      squ.perimeterFunc()
      squ.areaFunc()
      squ.showResult()
    elif row[0] == "pentagon":
      pent = Pentagon(row[1])
      pent.perimeterFunc()
      pent.areaFunc()
      pent.showResult()
    else:
      print("Invalid Shape. Error.")

```

```python
"""
   2. Given a raw string of basic html with no formatting such as:
      <html><body><div><a></a></div></body></html>
      
      Without using a parsing library, write a program to parse the input and produce the following output:
      <html>
        <body>
          <div>
            <a>
            </a>
          </div>
        </body>
      </html>
      
      You do not need to handle singleton or self-closing tags. 
      
      If the html is not valid (i.e., there are
      mismatched or missing tags), print an error. Thus, the following two strings should produce errors:
      <html><body><div></a></body></html>
      <html><body><div><a></div></a>

"""
from collections import deque
import re
def parseHtml(htmlstr):
  splitStr = re.split('(<[^>]*>)',htmlstr)[1::2] #slice [start:stop:step] since this regex will create empty ''
  #print(splitStr)
  count = 0

  stack = deque()
  for tag in splitStr:
    if tag[1] != "/":
      stack.append(tag)
      for j in range(count): print(end=" ")
      #print(end="\t"*count)
      count += 2
    else:
      splitTag = tag.replace('/', '') 
      popTag = stack.pop()
      if splitTag != popTag:
        print("There is a parsing error")
        return
      count -= 2
      for j in range(count): print(end=" ")
      #print(end="\t"*count)
      
    print(tag)



  return

parseHtml("<html><body><div><a></a></div></body></html>")
print("--------------")
parseHtml("<html><body><div><a></div></a>")
print("--------------")
parseHtml("<html><body><div><a></div></a>")
```

```python
"""
  3. Given an array of letters [a, b, c, d, e, f], write a recursive function that outputs the following
  structure:
  <a>
    <b>
      <c>
        <d>
          <e>
            <f>
            </f>
          </e>
        </d>
      </c>
    </b>
  </a>
"""
def printHtml(tags, i):
    if i >= len(tags):
        return
   
    print("\t"*i + "<" + tags[i] + ">")
    printHtml(tags, i + 1)
    print("\t"*i + "</" + tags[i] + ">")

tags = ['a','b','c','d','e']
printHtml(tags, 0)
```
